```python
song_data = {
    'name':[],
    'id':[],
    'album_type':[],
    'name':[],
    'track_number':[],
    'duration_ms':[],
    'explicit':[],
    'popularity':[],
}
for track in popular_playlist:
    name = track['name']
    song_data['name'].append(name)
    print(name)
    
    id = track['id']
    song_data['id'].append(id)
    print(id)
 
    album_type = track['album']['album_type']
    song_data['album_type'].append(album_type)
    print(album_type)
    
    name = track['album']['name']
    song_data['name'].append(name)
    print(name)
    
    track_number = track['track_number']
    song_data['track_number'].append(track_number)
    print(track_number)
    
    duration_ms = track['duration_ms']
    song_data['duration_ms'].append(duration_ms)
    print(duration_ms)
    
    explicit = track['explicit']
    song_data['explicit'].append(explicit)
    print(explicit)
    
    popularity = track['popularity']
    song_data['popularity'].append(popularity)
    print(popularity)
    
    print('*************************************')
```
